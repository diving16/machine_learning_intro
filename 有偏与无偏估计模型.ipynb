Population Y:
Mean(Y)=$\mu =\frac{1}{N}\sum_{i=1}^{N}Y_{i}$

Var(Y)=$\sigma=\frac{1}{N}\sum_{i=1}^N(Y_i-\mu)^2$

Sample X:
Mean(X)=$\overline X =\frac{1}{n}\sum_{j=1}^{n}X_{j}$

Var(X)=$s^2=\frac{1}{n-1}\sum_{i=1}^n(X_j-\overline X)^2$

```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
figsize(15,5)
import pandas as pd
import numpy as np

np.random.seed(42)
# The population N's size is 100000
N=100000
population=pd.Series(np.random.randint(1,11,N))
```

```python
samples={}
# The size is each sample
n=30
# The number of sample
num_samples=500
for i in range(num_samples):
    samples[i]=population.sample(n).reset_index(drop=True)
samples=pd.DataFrame(samples)
print(samples)
```

```python
# Delta degree of freedom ddof=0 divided by n,ddof=1 divided by n-1
biased_samples=samples.var(ddof=0).to_frame()
biased_samples
```

```python
biased_samples=biased_samples.expanding().mean()
biased_samples
```

```python
biased_samples.columns=["biased var estimate"]
biased_samples
```

```python
biased_samples.plot()
```

```python
unbiased_samples=samples.var(ddof=1).to_frame()
unbiased_samples=unbiased_samples.expanding().mean()
unbiased_samples.columns=["unbiased var estimate"]
```

```python
ax=unbiased_samples.plot()
biased_samples.plot(ax=ax)
real_population_var=pd.Series(population.var(ddof=0),index=samples.columns)
real_population_var.plot()
```
